```python
import threading
import winsound
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

import cv2
import imutils

SMTP_SERVER = "smtp.gmail.com"
SMTP_PORT = 465
SENDER_EMAIL = "testproject9116@gmail.com"
SENDER_PASSWORD = "kloy sfuf shtu cfqz"
RECIPIENT_EMAIL = "pramebadde@gmail.com"
```

```python
capt = cv2.VideoCapture(0, cv2.CAP_DSHOW)

capt.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
capt.set(cv2.CAP_PROP_FRAME_HEIGHT,480)
```

```python
_, frame = capt.read()
frame = imutils.resize(frame, width=500)
frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
frame = cv2.GaussianBlur(frame, (21, 21), 0)
```

```python
alarm = False
alarm_mode = False
alarm_counter = 0
```

```python
def send_email_alert():
    subject = "Alarm Triggered"
    body = "The motion alarm has been triggered. Please check the camera feed."

    message = MIMEMultipart()
    message["From"] = SENDER_EMAIL
    message["To"] = RECIPIENT_EMAIL
    message["Subject"] = subject
    message.attach(MIMEText(body, "plain"))

    try:
        with smtplib.SMTP_SSL(SMTP_SERVER, SMTP_PORT) as server:
            server.login(SENDER_EMAIL, SENDER_PASSWORD)
            server.sendmail(SENDER_EMAIL, RECIPIENT_EMAIL, message.as_string())
        print("Alert email sent.")
    except Exception as e:
        print(f"Failed to send alert email: {e}")


def tig_alarm():
    global alarm
    for _ in range(5):
        if not alarm:
            break
        print("Alarm!")
        winsound.Beep(2500, 1000)
    alarm = False
    send_email_alert()
```

```python
while True:
    _, frame2 = capt.read()
    frame2 = imutils.resize(frame2, width=500)

    if alarm_mode:
        frame_gray = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
        frame_gray = cv2.GaussianBlur(frame_gray, (5, 5), 0)

        diff = cv2.absdiff(frame_gray, frame)
        thresh = cv2.threshold(diff, 25, 255, cv2.THRESH_BINARY)[1]
        frame = frame_gray

        if thresh.sum() > 10000:
            alarm_counter += 1
        else:
            if alarm_counter > 0:
                alarm_counter -= 1

        cv2.imshow("Cam_L", thresh)
    else:
        cv2.imshow("Cam_L", frame)

    if alarm_counter > 20:
        if not alarm:
            alarm = True
            threading.Thread(target=tig_alarm).start()

    key = cv2.waitKey(30)
    if key == ord("a"):
        alarm_mode = not alarm_mode
        alarm_counter = 0
    if key == ord("q"):
        alarm_mode = False
        break

capt.release()
cv2.destroyAllWindows()

 
```
